# **За варіантом завдання для побудови графікі двох функцій в одній системі координат**

image.png

```
x = np.arange(-2*np.pi,2*np.pi,0.01) 
y1 = (x ** 3 - x ** 2 - x -1)
y2 = np.exp(x) - np.cos(x)
plt.plot(x, y1, label = r'$f(x)=x**3 - x**2 - x - 1$')
plt.plot(x, y2, color = 'green', label = r'$f(x)=e**x-cos(x)$')
plt.title('Графики функций: вариант 13')
plt.legend()
plt.show()
```

```
#Інше представлення функцій з іншим діапозоном 
x = np.linspace(0, 2 *np.pi, 100)
y1 = (x ** 3 - x ** 2 - x -1)
y2 = np.exp(x) - np.cos(x)
plt.plot(x, y1, 'm-.', label = r'$f(x)=x**3 - x**2 - x - 1$')
plt.plot(x, y2, color='#87a3cc', linestyle='--', label = r'$f(x)=e**x-cos(x)$')
plt.title('Графики функций: вариант 13')
plt.show()
```

# **Побудова діаграм за наданими даними**

```
#Стовпчикова діаграма
labels = ["июль", "август", "сентябрь", "октябрь", "ноябрь", "декабрь"]
values = [34, 67, 12, 7, 41, 39]
width = 0.3
x = np.arange(len(labels))
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, values, width, label='Полученные значения', color = 'r')
ax.set_title('Объем продаж ПК фирмой')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

```
#Горизонтальна стовпчикова діаграма зі зміненими параметрами фігури та кольорами
labels = ["июль", "август", "сентябрь", "октябрь", "ноябрь", "декабрь"]
values = [34, 67, 12, 7, 41, 39]
dpi = 80
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
mpl.rcParams.update({'font.size': 9})
plt.title('Объем продаж ПК фирмой')
ax = plt.axes()
ax.xaxis.grid(True, zorder = 1)
xs = range(len(labels))#
plt.barh([x + 0.05 for x in xs], values, height = 0.2, color = 'orange', alpha = 0.7, label = 'Полученные значения', zorder = 2)
plt.yticks(xs, labels, rotation = 10)
plt.legend(loc='upper right')
fig.savefig('barshoris.png')
```

```
#Ріе-діаграма з кольоровою гамою
import seaborn as sb
labels = ["июль", "август", "сентябрь", "октябрь", "ноябрь", "декабрь"]
sizes = [34, 67, 12, 7, 41, 39]
explode = (0, 0.1, 0, 0, 0.1, 0)
colors = sb.color_palette("BrBG", 7)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=100)
ax1.axis('equal')
plt.show()
```
